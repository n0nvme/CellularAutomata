```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format='retina'
```

```python
def tor(x):
    if x < 0:
        return x + count
    else:
        return x % count
```

```python
def cellularMath(y1, y2, y3, rule=22):
    rule_bin = str(bin(rule))[2:]
    while len(rule_bin) != 8:
        rule_bin = '0' + rule_bin
    rule_bin = list(map(int, rule_bin))
    
    if y1 == 1 and y2 == 1 and y3 == 1:
        return rule_bin[0]
    elif y1 == 1 and y2 == 1  and y3 == 0:
        return rule_bin[1]
    elif y1 == 1 and y2 == 0 and y3 == 1:
        return rule_bin[2]
    elif y1 == 1 and y2 == 0 and y3 == 0:
        return rule_bin[3]
    elif y1 == 0 and y2 == 1 and y3 == 1:
        return rule_bin[4]
    elif y1 == 0 and y2 == 1 and y3 == 0:
        return rule_bin[5]
    elif y1 == 0 and y2 == 0 and y3 == 1:
        return rule_bin[6]
    elif y1 == 0 and y2 == 0 and y3 == 0:
        return rule_bin[7]
```

```python
count = 256
field = np.zeros((count))
field[count // 2] = 1
field1 = np.zeros((count))
result = np.zeros((count, count))
```

```python
for k in range(count):# количество строк(поколений автомата)
    result[k] = field
    for i in range(count):
        field1[i] = cellularMath(field[tor(i-1)], field[tor(i)], field[tor(i + 1)], 15)
    field = field1
    field1 = np.zeros((count))
```

```python
plt.figure(figsize=(10, 10))
plt.xticks([])
plt.yticks([])
plt.imshow(result, cmap='Greys')
```
